### Univariate Analysis

```
sns.displot(am['normalized_losses'], kde=True)
```

```
sns.distplot(am['number_of_cylinders'])
```

### Bivariate Analysis

```
sns.jointplot(am['engine_size'],am['horsepower'])
```

```
sns.pairplot(am[['normalized_losses','engine_size','horsepower']])
```

### Categorical and Numerical Value together

```
plt.figure(figsize=(12,5))
sns.swarmplot(am['fuel_type'],am['horsepower'])
```

```
sns.boxplot(am['number_of_doors'],am['horsepower'],hue=am['fuel_type'])
```

```
sns.barplot(am['body_style'],am['horsepower'])
```

```
sns.barplot(am['body_style'],am['horsepower'],hue=am['engine_location'])
```

```
sns.pointplot(am['fuel_system'],am['horsepower'],hue=am['number_of_doors'])
```

```
sns.pointplot(am['fuel_system'],am['horsepower'],hue=am['number_of_cylinders'])
```

```
sns.pointplot(am['fuel_system'],am['horsepower'],hue=am['fuel_system'])
```

```
sns.factorplot(x='fuel_type',y='horsepower',hue='number_of_doors',col='engine_location',data=am,kind='violin')
```
